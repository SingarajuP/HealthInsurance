```python
# Required packages
import numpy as np
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns

%matplotlib inline 
```

```python
#loading train and test data into dataframe
df_train=pd.read_csv("train_Df64byy.csv",delimiter=',')
df_test=pd.read_csv("test_YCcRUnU.csv",delimiter=',')
```

```python
df_train.head()
```

Imputing missing values in both train and test dataframe

```python
df_train['Health Indicator'].fillna(df_train['Health Indicator'].mode()[0], inplace=True)
df_train['Holding_Policy_Duration'].fillna(0, inplace=True)
df_train['Holding_Policy_Type'].fillna(0, inplace=True)

df_test['Health Indicator'].fillna(df_test['Health Indicator'].mode()[0], inplace=True)
df_test['Holding_Policy_Duration'].fillna(0, inplace=True)
df_test['Holding_Policy_Type'].fillna(0, inplace=True)
```

```python
df_train['Holding_Policy_Type'] = df_train['Holding_Policy_Type'].astype(str)
df_train['Reco_Policy_Cat'] = df_train['Reco_Policy_Cat'].astype(str)
df_train['Region_Code'] = df_train['Region_Code'].astype(str)
```

```python
print("Train dataset:\n", df_train.isna().sum())
print("Test dataset:\n",df_test.isna().sum())
```

```python
df_train.dtypes,df_test.dtypes
```

```python
X=df_train.drop(['ID','Response'],axis=1)
y=df_train['Response']
```

```python
X.shape, X.dtypes, y.shape, y.dtypes
```

```python
cat_features=np.where(X.dtypes =='object')[0]
cat_features
```

### Modeling

```python
from sklearn.model_selection import train_test_split
import sklearn.model_selection as skms
from catboost import CatBoostClassifier
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.metrics import accuracy_score
```

```python
X_train , X_valid, y_train, y_valid = train_test_split(X,y, 
                                                       test_size=0.2, 
                                                       random_state=1, 
                                                       stratify=y)
```

The Evaluation metric is ROC_AUC_SCORE

```python
params = {
    'cat_features': cat_features,
    'eval_metric': 'AUC',
    'random_seed': 21,
    'n_estimators': 10000,
}
```

```python
model=CatBoostClassifier(**params,early_stopping_rounds=100)
```

```python
model.fit(X_train,y_train)
#model.fit(X_train, y_train, early_stopping_rounds=100,verbose=100,cat_features=cat_features)
```

```python
ypred_cat=model.predict_proba(X_valid)[:,1]
```

```python
print("The ROC_AUC_SCORE with CATBOOST is :",roc_auc_score(y_valid, ypred_cat))
```
